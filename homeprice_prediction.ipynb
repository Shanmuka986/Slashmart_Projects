```python
def outlier_condition(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (df[col] >= lower) & (df[col] <= upper)

# Start with all True mask
mask = pd.Series(True, index=df.index)

# Select only numerical columns for outlier detection, excluding the one-hot encoded zipcode columns
numerical_cols = df.select_dtypes(include=['number']).columns
outlier_cols = [col for col in numerical_cols if not col.startswith('zipcode_')]


# Combine masks
for col in outlier_cols:
    mask = mask & outlier_condition(df, col)

# Apply once
df = df[mask]
```

```python
#boxplot vertical


for col in df:
    plt.figure(figsize=(8, 4))
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#onehot encoding on zipcode
df = pd.get_dummies(df, columns=['zipcode'], drop_first=True)
```

```python
df.head()
```

```python
#linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#splitiing the data into training and testing
x=df.drop(columns=['price'])
y=df['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model=LinearRegression()
model.fit(x_train, y_train)

model.score(x_test,y_test)
```

```python
#gradient boosting
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(
    n_estimators=400,
    learning_rate=0.1,
    max_depth=5,
    min_samples_split=2,
    random_state=2
)
model.fit(x_train, y_train)
print(model.score(x_test, y_test))
```
